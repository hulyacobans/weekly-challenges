```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Using example data from Colab

```
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

Missing Values Chart

```
plt.figure(figsize=(15,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
plt.show()
```

Correlation Chart

```
df_kor = df.corr()
plt.figure(figsize=(15,15))
sns.heatmap(df_kor, vmin=-1, vmax=1, cmap="viridis", annot=True, linewidth=0.1)
```

Joint Plot

```
sns.set(color_codes=True)
sns.jointplot(x=df['median_income'],
              y=df['median_house_value'],
              data=df,
              kind='kde')

plt.plot()
```

Pair Plot

```
sns.pairplot(df)
```

Scatter Plot

```
sns.scatterplot(data=df, x="population", y="housing_median_age")
```

```
sns.scatterplot(data=df, x="population", y="housing_median_age", hue="median_income")
```
